```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
relatorio_de_vendas_df = pd.read_csv("Relatório de vandas.csv", sep=';')
lista_de_colunas = ['Código','Data Abertura','Cliente','Data Fechamento','Total']
relatorio_de_vendas_df = relatorio_de_vendas_df[lista_de_colunas]
relatorio_de_vendas_df = relatorio_de_vendas_df[relatorio_de_vendas_df['Cliente'] != None]

display(relatorio_de_vendas_df)
relatorio_de_vendas_df.to_csv('relatorio_de_vendas_df.csv',sep=';', encoding='utf-8', index=False)
```

```python
historico_vendidas_df = pd.read_csv('Historico_Itens_Vendidos.csv',sep=';')
historico_vendidas_df = historico_vendidas_df.rename(columns={'Valor Un. Item': 'temp'})
historico_vendidas_df = historico_vendidas_df.rename(columns={'Valor. Tot. Item': 'Valor Un. Item'})
historico_vendidas_df = historico_vendidas_df.rename(columns={'temp': 'Valor. Tot. Item'})
historico_vendidas_df = historico_vendidas_df.drop(columns=['Data/Hora Item','Tipo Prod','Núm. Mesa/Com.','Stat. Ped.','Tipo Ped.'])
display(historico_vendidas_df)
historico_vendidas_df.to_csv('historico_vendidas_df.csv',sep=';', encoding='utf-8', index=False)
historico_vendidas_df.info()
```

```python
movimentacao_estoque_df = pd.read_csv('movimentacoes-estoque.csv',sep=';')
movimentacao_estoque_df = movimentacao_estoque_df[movimentacao_estoque_df['Operação'] == 'Saída']
movimentacao_estoque_df = movimentacao_estoque_df[movimentacao_estoque_df['Un. Med.'] != 'UN']
movimentacao_estoque_df = movimentacao_estoque_df.drop(columns=['Observações','Est. Anterior','Est. Final','Usuário','E/S'])
display(movimentacao_estoque_df)
movimentacao_estoque_df.to_csv('movimentacao_estoque_df.csv',sep=';', encoding='utf-8',index=False)
```
